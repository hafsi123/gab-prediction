**adding week-days feature[1,2,3,4,5,6,7]**

```python
import pandas as pd

# create a dictionary to map day of the week to a number
days_map = {
    'Monday': 1,
    'Tuesday': 2,
    'Wednesday': 3,
    'Thursday': 4,
    'Friday': 5,
    'Saturday': 6,
    'Sunday': 7
}

# create a new column 'nbr_day' based on the day of the week in column 'dateT'
df['nbr_day'] = pd.to_datetime(df['dateT']).dt.day_name().map(days_map)
```

```python
df['nbr_day'].unique()
```

**adding working days and weekends feature**


*   work_day ['monday','tuesday','wednesday','thursday','friday'] =0
*   weekend ['saturday','sunday'] =1







```python
def is_working_day(day):
    if day in [1, 2, 3, 4, 5]:
        return 0 # working day
    else:
        return 1 # weekend day

df['working_day'] = df['nbr_day'].apply(is_working_day)
```

```python
df['working_day'].unique()
```

**setting temperature feature for day/night shifts**


*   day == [(TEMPERATURE_MORNING_C,TEMPERATURE_NOON_C'].mean()
*   night == [TEMPERATURE_EVENING_C]



```python
df['temperature'] = df.apply(lambda row: (row['TEMPERATURE_MORNING_C'] + row['TEMPERATURE_NOON_C'])/2 if row['Day_Night'] == 'DAY' else row['TEMPERATURE_EVENING_C'], axis=1)
```

```python
df['dateT'] = pd.to_datetime(df['dateT'])

# Add a new column 'Month' to store the month of each date
df['Month'] = df['dateT'].dt.month

# Calculate the mean temperature for each month
monthly_avg_temp = df.groupby('Month')['temperature'].mean()

# Print the monthly average temperature
print(monthly_avg_temp.tolist())
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
data=df
# select the columns for the feature matrix X and the target variable y
X = df
y = data['Total_Amount']


categorical_cols = ['GAB', 'Day_Night', 'Region']
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])

# define the column transformer to handle numerical columns
numerical_cols = ['Code_Postal', 'Nbr_Transaction', 'MAX_TEMPERATURE_C',
                  'MIN_TEMPERATURE_C', 'WINDSPEED_MAX_KMH', 'TEMPERATURE_MORNING_C',
                  'TEMPERATURE_NOON_C', 'TEMPERATURE_EVENING_C', 'PRECIP_TOTAL_DAY_MM',
                  'HUMIDITY_MAX_PERCENT', 'VISIBILITY_AVG_KM', 'PRESSURE_MAX_MB',
                  'CLOUDCOVER_AVG_PERCENT', 'HEATINDEX_MAX_C', 'DEWPOINT_MAX_C',
                  'WINDTEMP_MAX_C', 'TOTAL_SNOW_MM', 'UV_INDEX', 'SUNHOUR',
                  'TEMPERATURE_NIGHT_C', 'avg_trans']
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median'))])

# combine the transformers with ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols)])

# create a pipeline that includes the preprocessor and LinearRegression model
model = Pipeline(steps=[('preprocessor', preprocessor),
                        ('regressor', LinearRegression())])

# fit the model
model.fit(X, y)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_squared_error, r2_score


# Split the data into training and testing sets

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)



# Train the Random Forest Regression model
rf_reg = RandomForestRegressor(n_estimators=100, random_state=42)
rf_reg.fit(X_train, y_train)

# Make predictions on the testing set
y_pred = rf_reg.predict(X_test)

# Evaluate the model's performance
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)
print("R-squared score:", r2)
print("Root Mean Squared Error:", rmse)
```
